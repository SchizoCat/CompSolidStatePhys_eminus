# Exercise 1: Simple SCF Calculations with Crystals (10 points)

A self-consistent field (scf) calculation is one where the Kohn-Sham system of equations given by
\begin{equation}
\Bigg[-\frac12 \nabla^2 + \int \frac{n(\vec{r}')}{\vert \vec{r}-\vec{r}' \vert} d\vec{r}'+V_{XC}+V_{ext} \Bigg]\phi_i(\vec{r})=\varepsilon_i\phi_i(\vec{r}).
\end{equation}
are solved in an iterative manner for $\{\phi_i(\vec{r})\}$ for a given $n(\vec{r})$. In order to achieve this, Kohn-Sham orbitals are expanded in terms of plane-waves and cast into a linear algebra system of equations. The solution is conducted iteratively where at every step, the density $n(\vec{r})$ is updated using the new solutions, and the potential is recalculated, which is then ready for the next step.

In this exercise, you will learn to set the variables for a typical self-consistent field (scf) calculation in the context of a crystal. Once the variables are set and the simulation box is created, you will run the scf calculation and analyze the results. We will solve the same problem twice: once in the primitive unit cell and one in the conventional cell.

```python
### 1. Diamond C using the primitive unit cell
## Enter the parameters of the SCF calculation
a =   3.567              ## Lattice constant in A
a = ang2bohr(a)
atom =  "C"              ## Atom symbol
lattice = "diamond"      ## Lattice type
ecut = 20                ## Kinetic energy cutoff
kmesh= [4,4,4]           ## K-point mesh (MP grid)
#basis=None               ## Basis atoms
bands=None               ## Bands
smearing=0               ## Smearing for better BZ integration
magnetization=None       ## Initial total magnetic moment (for spin-pol calculations)
#verbose=None             ## Verbosity level

## Call the Cell constructor to make a Cell object
cell = Cell(a=a,atom=atom, lattice=lattice, ecut=ecut, kmesh=kmesh)

## Run the SCF calculation
scf = SCF(cell)
scf.run()

## Print the total energy
print(scf.energies)
## View the atoms
scf.view()

## View the density
from eminus.extras import view
view(scf, plot_n=True, surfaces=20, percent=40)
```
